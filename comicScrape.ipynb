```python
import requests
from bs4 import BeautifulSoup
import urllib.parse

base_url = "https://www.comics.org"
search_url = "https://www.comics.org/search/advanced/process/?target=issue_cover&method=icontains&logic=False&type=19&order1=series&order2=date&start_date=01%2F01%2F1900&end_date=01%2F01%2F1961&country=us&page={}"

# Iterate over each page
for page in range(1, 11):  # Change the range as per the number of pages you want to scrape
    # Generate the search URL for the current page
    page_url = search_url.format(page)
    
    # Send a GET request to the search URL
    response = requests.get(page_url)
    soup = BeautifulSoup(response.text, "html.parser")

    # Find all the comic links on the page
    comic_links = soup.find_all("a", class_="small button")

    # Iterate over each comic link
    for link in comic_links:
        comic_url = urllib.parse.urljoin(base_url, link["href"])
        response = requests.get(comic_url)
        comic_soup = BeautifulSoup(response.text, "html.parser")
        
        # Find the image element
        image_element = comic_soup.find("img", class_="comic-image")

        if image_element:
            image_url = urllib.parse.urljoin(base_url, image_element["src"])
            print("Image URL:", image_url)
```
